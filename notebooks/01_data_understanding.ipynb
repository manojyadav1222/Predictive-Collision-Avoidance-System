## Datetime Conversion

The time-related columns are currently stored as strings (object type).

Converting them to datetime format enables extraction of useful temporal features such as hour, month, weekday, and accident duration.

```python
sample_df["Start_Time"] = pd.to_datetime(
    sample_df["Start_Time"],
    format="mixed"
)

sample_df["End_Time"] = pd.to_datetime(
    sample_df["End_Time"],
    format="mixed"
)

sample_df["Weather_Timestamp"] = pd.to_datetime(
    sample_df["Weather_Timestamp"],
    format="mixed"
)
```

```python
sample_df[
    ["Start_Time",
     "End_Time",
     "Weather_Timestamp"]
].dtypes
```

### Datetime Parsing Challenge

The dataset contained multiple timestamp formats. Some records included nanosecond precision while others did not.

To ensure consistent datetime conversion, Pandas' `format="mixed"` option was used, allowing automatic inference of different timestamp formats.

```python
# Hour
sample_df["Hour"] = sample_df["Start_Time"].dt.hour

# Month
sample_df["Month"] = sample_df["Start_Time"].dt.month

# Day of Week
sample_df["DayOfWeek"] = sample_df["Start_Time"].dt.dayofweek

# Year
sample_df["Year"] = sample_df["Start_Time"].dt.year

# Weekend Flag
sample_df["Is_Weekend"] = (
    sample_df["DayOfWeek"] >= 5
).astype(int)

# Accident Duration
sample_df["Accident_Duration_Minutes"] = (
    sample_df["End_Time"] -
    sample_df["Start_Time"]
).dt.total_seconds() / 60
```

## Outlier Treatment

The engineered feature `Accident_Duration_Minutes` contained extreme values, including incidents lasting several years.

Since such durations are unrealistic for traffic accidents and likely represent data quality issues, records with accident durations greater than 1440 minutes (24 hours) were removed.

Only 2,321 records (approximately 0.46% of the dataset) were removed, preserving over 99.5% of the data.
