# Data Preprocessing

## Objective

The goal of this notebook is to prepare the dataset for machine learning by:

- Removing irrelevant features
- Handling missing values
- Converting data types
- Creating model-ready data

```python
sample_df.info()
```

```python
sample_df.isnull().sum().sort_values(ascending=False)
```

```python
columns_to_drop = [
    "ID",
    "Country",
    "Turning_Loop",
    "End_Lat",
    "End_Lng",
    "Description",
    "Street",
    "Zipcode"
]
```

```python
sample_df=sample_df.drop(
    columns=columns_to_drop
)
```

```python
sample_df.shape
```

```python
sample_df.columns.tolist()
```

```python
missing_report = pd.DataFrame({
    "Missing Values": sample_df.isnull().sum(),
    "Percentage": round(
        sample_df.isnull().sum() / len(sample_df) * 100,
        2
    )
})

missing_report = missing_report.sort_values(
    by="Percentage",
    ascending=False
)

missing_report[missing_report["Missing Values"] > 0]
```

```python
numerical_cols = sample_df.select_dtypes(
    include=["int64", "float64"]
).columns

numerical_cols
```

```python
categorical_cols = sample_df.select_dtypes(
    include=["object"]
).columns

categorical_cols
```

```python
missing_report[missing_report["Missing Values"] > 0]
```

```python
numerical_missing = [
    "Precipitation(in)",
    "Wind_Chill(F)",
    "Wind_Speed(mph)",
    "Visibility(mi)",
    "Humidity(%)",
    "Temperature(F)",
    "Pressure(in)"
]
```

```python
for col in numerical_missing:
    sample_df[col].fillna(
        sample_df[col].median(),
        inplace=True
    )
```

```python
categorical_missing = [
    "Wind_Direction",
    "Weather_Condition",
    "Airport_Code",
    "Timezone",
    "City",
    "Sunrise_Sunset",
    "Civil_Twilight",
    "Nautical_Twilight",
    "Astronomical_Twilight"
]
```

```python
for col in categorical_missing:
    sample_df[col].fillna(
        sample_df[col].mode()[0],
        inplace=True
    )
```

```python
sample_df["Weather_Timestamp"].fillna(
    sample_df["Start_Time"],
    inplace=True
)
```

## Missing Value Treatment

Missing values were handled using the following strategy:

### Numerical Features
Missing values were replaced using the median value of each feature.

### Categorical Features
Missing values were replaced using the most frequent category (mode).

### Datetime Features
Missing values in Weather_Timestamp were replaced using Start_Time.
